## Data Preparation

### Data Cleaning

### The Numbers

We start with the `numbers_df` data frame

```python
numbers_df
```

```python
numbers_df.info()
```

We note that the datatype above is all object and would like to convert some of the columns to integers so we can carry out some mathematical operations. Also we not that the `production_budget`, `domestic_gross` and `worldwide_gross` columns have characters that are not integers ('$',','), which will get in the way of converting the data type.

```python
# We start by removing the comas and dollar sign in the 3 columns mentioned above
numbers_df['production_budget'] = numbers_df['production_budget'].str.replace(',','').str.replace('$','')
numbers_df['domestic_gross'] = numbers_df['domestic_gross'].str.replace(',','').str.replace('$','')
numbers_df['worldwide_gross'] = numbers_df['worldwide_gross'].str.replace(',','').str.replace('$','')
numbers_df
```

We see now the columns have clean numbers that can be easily converted to integers.

```python
# Convert the three columns from object to integers
numbers_df = numbers_df.astype({'production_budget':'int64','domestic_gross':'int64', 'worldwide_gross':float})
# Check for confirmation
numbers_df.info()
```

With this done we can now carry out mathematical operations on the columns.

We start by making a new column `total_gross` which represents the total amount of money the movie earned.

```python
# We make a new column with the total_gross
numbers_df['total_gross'] = numbers_df['domestic_gross'] + numbers_df['worldwide_gross']
numbers_df
```

Now that we know the total money each movie makes and the cost of production, we can now easily calculate the `profit`.

```python
# We make the profit column
numbers_df['profit'] = numbers_df['total_gross'] - numbers_df['production_budget']
numbers_df
```

Furthermore I want to compare the numbers of each movie on an yearly basis. To do this I would need to make a column containing the year of release that is easier to work with than `release_date` which combines date,month and year.

```python
# I separete the release_date column into month, date and year respectively
numbers_df['month'] = pd.to_datetime(numbers_df['release_date'], format='%b %d, %Y').dt.month
numbers_df['date'] = pd.to_datetime(numbers_df['release_date'], format='%b %d, %Y').dt.day
numbers_df['year'] = pd.to_datetime(numbers_df['release_date'], format='%b %d, %Y').dt.year
numbers_df
```

I proceed to make a smaller dataframe with just the columns I am interested in working with. 

These columns are:
 * `production_budget`
 * `profit`
 * `year`

```python
# Make smaller dataframe
prod_profit_df = numbers_df[['movie','production_budget','profit','year']]
prod_profit_df
```
